## Applying Index Mask

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
countries_df = pd.read_csv('countries_of_the_world.csv', decimal= ',')
countries_df.head(2)
```

```python
birthrates = countries_df['Birthrate'].values
```

```python
type(birthrates)
```

```python
birthrates.shape
```

```python
plt.plot(birthrates)
plt.show()
```

```python
np.median(birthrates)
```

```python
birthrates
```

```python
birthrates = birthrates[~np.isnan(birthrates)]
```

```python
birthrates.shape
```

```python
np.median(birthrates)
```

```python
np.count_nonzero(birthrates[birthrates<20])
```

```python
np.sort(birthrates)[0:20]
```
